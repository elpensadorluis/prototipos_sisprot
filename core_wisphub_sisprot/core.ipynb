```python
import requests
import numpy as np
import pandas as pd
import json as json
```

```python
def peticion(url, key):
    b = ''
    while True:
        response2 = requests.get(url, headers=key)
        respuesta2 = response2.json()
        
        api_url = respuesta2['next']
        a = np.array(respuesta2['results'])
        if len(b) <= 0:
            b = a
        else:
            b = np.concatenate((a,b))
        # the below condition will tell
        # the loop to stop
        print(f"{len(b):4} de {str(respuesta2['count'])} : {len(b)/respuesta2['count']:.2%}")
        
        if len(b) >= respuesta2['count']:
            break
    
    return b
```
